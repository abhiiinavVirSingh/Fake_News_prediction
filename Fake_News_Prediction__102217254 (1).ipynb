```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

data_true = pd.read_csv('True.csv')
data_true.head()

```python
data_fake= pd.read_csv('Fake.csv')
data_fake.head()
```

```python
data_fake['class']=0
data_true['class']=1
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480,23470,-1):
    data_fake.drop([i],axis=0,inplace=True)


data_true_manual_testing = data_true.tail(10)
for i in range(21416,21406,-1):
    data_true.drop([i],axis=0,inplace=True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing['class']=0
data_true_manual_testing['class']=1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data= pd.concat([data_fake,data_true],axis=0)
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data = data.drop(["title", "subject","date"], axis = 1)
```

```python
data.isnull().sum()
```

```python
# We are doing random shuffling since the true and fake news are together seperately after merging
data = data.sample(frac=1)
data.reset_index(inplace=True)
data.drop(['index'], axis=1, inplace=True)
```

```python
data.head()
```

```python

def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ", text)
    text = re.sub('https?//\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
data['text']= data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(C=0.1)
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from scipy.stats import uniform

# Define a smaller parameter distribution
param_dist = {
    'C': uniform(0.01, 1),           # Reduced range for regularization
    'penalty': ['l1', 'l2'],         # L1 and L2 regularization
    'solver': ['liblinear'],         # Fixed solver to reduce options
    'max_iter': [100, 200]           # Fewer choices for max iterations
}

# Initialize the model
LR = LogisticRegression()

# Apply RandomizedSearchCV with fewer iterations
random_search = RandomizedSearchCV(
    LR,
    param_distributions=param_dist,
    n_iter=10,  # Reduced to 10 candidates
    cv=5,
    scoring='accuracy',
    random_state=42,
    verbose=1
)
random_search.fit(xv_train, y_train)

# Best parameters and model
print("Best Parameters:", random_search.best_params_)
```

```python
# Define output_label function
def output_label(n):
    if n == 0:
        return "Fake news"
    elif n == 1:
        return "Not Fake News"
    else:
        return "Unknown"

# Define manual_testing function
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)

    # Predictions using available models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)

    # Print predictions for the models used
    print("\n\nLR Prediction: {} \nDT Prediction: {}".format(
        output_label(pred_LR[0]),
        output_label(pred_DT[0])
    ))

# Input for manual testing
news = str(input("Enter a news article to test: "))
manual_testing(news)

```
